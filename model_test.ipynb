```python
from torch.utils.data import Dataset, DataLoader
import tqdm
import torch
import random

def mask_tokens(inputs, vocab, mask_id, prob=0.15):
    """ Prepare masked tokens inputs/labels for masked language modeling: 80% MASK, 10% random, 10% original. """
    labels = inputs.clone()

    masked_indices = torch.bernoulli(torch.full(labels.shape, prob)).bool()
    labels[~masked_indices] = -1  
    indices_replaced = torch.bernoulli(torch.full(labels.shape, 0.8)).bool() & masked_indices
    inputs[indices_replaced] = mask_id

    # 10% of the time, we replace masked input tokens with random word
    """
    mask_indices20% ,random ,. 
    """
    indices_random = torch.bernoulli(torch.full(labels.shape, 0.5)).bool() & masked_indices & ~indices_replaced
    random_words = torch.randint(len(vocab)+1, labels.shape, dtype=torch.long)
    inputs[indices_random] = random_words[indices_random]

    return inputs, labels 

class OCRBDataset(Dataset):
    def __init__(self, data, vocab, mask_id):
        self.data = data
        self.vocab = vocab
        self.mask_id = mask_id
    
    def __len__(self):
        return self.data.shape[0]
    
    def __getitem__(self, item):
        inputs,label = mask_tokens(torch.from_numpy(self.data[item].toarray().astype(int)), self.vocab, self.mask_id)
        if len(inputs.shape)==3:
            inputs=torch.squeeze(inputs, dim=1)
            label=torch.squeeze(label, dim=1)
        output = {"ocrb_input": inputs,
                  "ocrb_label": label}
        # return {key: torch.tensor(value) for key, value in output.items()}
        return output
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

seq_len=135
#  15 
target_word_idx = 64
target_attention = attn[68745][target_word_idx]
target_attention

# 
fig, ax = plt.subplots(figsize=(12, 6))

#  x  y 
x_positions = np.arange(seq_len)
y_position = 1.0  #  y 
bar_base = 0.8  #  y 
# 
curve_amplitude = 0.3  # 
linewidth_scale = 40  # 
# 

for i, attn_weight in enumerate(np.array(target_attention)):
    if attn_weight > 0.01:  # 
        # 
        start = (target_word_idx, y_position)  # 
        end = (i, y_position)  # 
        control = ((target_word_idx + i) / 2, y_position + curve_amplitude)  # y 

        # 
        t = np.linspace(0, 1, 100)  # 
        bezier_x = (1 - t)**2 * start[0] + 2 * (1 - t) * t * control[0] + t**2 * end[0]
        bezier_y = (1 - t)**2 * start[1] + 2 * (1 - t) * t * control[1] + t**2 * end[1]

        # 
        ax.plot(
            bezier_x, bezier_y,
            alpha=min(attn_weight * 5, 1.0),  # 
            color="purple",
            linewidth=attn_weight * linewidth_scale  # 
        )
        
        ax.scatter(i, y_position, color="blue", s=10)  # 

        ax.bar(
            i,  #  x 
            attn_weight * 2,  # 
            color="blue",
            alpha=0.6,
            width=0.4,
            bottom=bar_base  #  bar_base y_others
        )
        
        ax.bar(
            i,  #  x 
            GAP_data[68745].toarray()[0][i] * 0.02,  # 
            color="blue",
            alpha=0.6,
            width=0.4,
            bottom=0.6  #  bar_base y_others
        )

# 
# ax.scatter(x_positions, [y_others] * seq_len, color="blue", s=50, label="Words")  # 
# 
ax.hlines(
    y=bar_base,  #  y 
    xmin=-0.5,  #  x 
    xmax=seq_len - 0.5,  #  x 
    colors="black",  # 
    linewidth=0.5,  # 
    linestyles="-"  # 
)
# 
ax.scatter([target_word_idx], [y_position], color="red", s=10, label="Target Word")  # 

# 
ax.set_xlim(-1, seq_len)  # x 
ax.set_ylim(0.5, 1.5)  # y 
# ax.set_xticks(x_positions)  # x 
ax.set_yticks([])  #  y 
ax.set_title(f"Attention Relationships for Target Word {target_word_idx} (Curved Connections)", fontsize=14)
ax.legend()

# 
plt.tight_layout()
plt.show()
```
